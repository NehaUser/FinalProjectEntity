# Code to Wrangle usdaHoney2 and make it usable for further Analysis.

# Load Libraries

```python
#Load Libraries

import pandas as pd
from scipy import stats
```

# Load Data

```python
# Loading the usdaHoney file

usdaHoney2 = pd.read_csv("../Data/usdaHoney2.csv")
```

```python
# Looking at the data and assigning to a new dataframe

usdaHoney2.head()
usdaHoneyDS = usdaHoney2
```

```python
# Looking at different Units of Measurements and converting into one unique UoM

usdaHoneyDS['Unit_Description'].unique()
```

# Data Wrangling

```python
# Changed the Units to Metric Tons

# Filtering the Commodities with non calculable UofM

usdaHoneyDS = usdaHoneyDS[usdaHoneyDS.Unit_Description != '(1000 HEAD)']
usdaHoneyDS = usdaHoneyDS[usdaHoneyDS.Unit_Description != '(MIL PCS)']

# Converting Bags to MT

usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(1000 60 KG BAGS)", "Value"] = (usdaHoneyDS.Value * 1000 * .001 * 60)
usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(1000 60 KG BAGS)", "Unit_Description"] = "(MT)"

# 
usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(MT, Net Weight)", "Unit_Description"] = "(MT)"
usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "MT", "Unit_Description"] = "(MT)"


# Converting 100 MT to MT

usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(1000 MT)", "Value"] = (usdaHoneyDS.Value * 1000)
usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(1000 MT)", "Unit_Description"] = "(MT)"

# Coverting Bales to MT

usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "1000 480 lb. Bales", "Value"] = (usdaHoneyDS.Value * 1000 * .00045 * 480)
usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "1000 480 lb. Bales", "Unit_Description"] = "(MT)"


# Converting CWE to MT

usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(1000 MT CWE)", "Value"] = (usdaHoneyDS.Value * 1000)
usdaHoneyDS.loc[usdaHoneyDS["Unit_Description"] == "(1000 MT CWE)", "Unit_Description"] = "(MT)"
```

```python
usdaHoneyDS1 = usdaHoneyDS
```

```python
# Picking commodities Categorywise and Wrangling it
# Category 1: Dairy and Poultry.

print(usdaHoneyDS1.Commodity_Description[usdaHoneyDS1.Category == 1].unique())
```

```python
# Deleting the Commodities with high multicollinearity or missing values

Cat1 = usdaHoneyDS1.groupby(['Commodity_Description','Year'], as_index=False).agg(max)
 
indexes = Cat1[(Cat1['Commodity_Description'] == 'Dairy, Butter')
               | (Cat1['Commodity_Description'] == 'Dairy, Cheese')
               | (Cat1['Commodity_Description'] == 'Poultry, Meat, Broiler')
               | (Cat1['Commodity_Description'] == 'Dairy, Dry Whole Milk Powder')
               | (Cat1['Commodity_Description'] == 'Dairy, Milk, Nonfat Dry')].index
  
Cat1.drop(indexes,inplace=True)

Cat1.head(20)
```

```python
Cat2 = Cat1
```

```python
# Wrangling the Data for Category2 : Fruits and Vegetables

Cat2 = Cat2[Cat2.Commodity_Description != 'Tangerines, Fresh']
Cat2 = Cat2[Cat2.Commodity_Description != 'Grapes, Table, Fresh']
Cat2 = Cat2[Cat2.Commodity_Description != 'Grapes']
Cat2 = Cat2[Cat2.Commodity_Description != 'Grapes, Fresh Table']
Cat2 = Cat2[Cat2.Commodity_Description != 'Cherries (Sweet&Sour), Fresh']
Cat2 = Cat2[Cat2.Commodity_Description != 'Peaches, Canned']
Cat2 = Cat2[Cat2.Commodity_Description != 'Strawberries, Frozen']
Cat2 = Cat2[Cat2.Commodity_Description != 'Prunes (Plums, Dried)']
Cat2 = Cat2[Cat2.Commodity_Description != 'Lemons, Fresh']
Cat2 = Cat2[Cat2.Commodity_Description != 'Peaches & Nectarines, Fresh']
Cat2 = Cat2[Cat2.Commodity_Description != 'Apple Juice, Concentrated']
Cat2 = Cat2[Cat2.Commodity_Description != 'Orange Juice']
Cat2 = Cat2[Cat2.Commodity_Description != 'Citrus, Other, Fresh']
Cat2 = Cat2[Cat2.Commodity_Description != 'Potato Products, Frozen']
Cat2 = Cat2[Cat2.Commodity_Description != 'Raisins']
```

```python
Cat2.head(20)
```

```python
print(Cat2.Commodity_Description[Cat2.Category == 2].unique())
```

```python
# Wrangling for category 3 : Grains 

Cat3 = Cat2

indexes3 = Cat3[Cat3['Commodity_Description'].str.match('Meal')].index
  
Cat3.drop(indexes3,inplace=True)

print(Cat3.Commodity_Description[Cat3.Category == 3].unique())

```

```python
# Wrangling for category 4 : Nuts and Oil 

Cat4 = Cat3

indexes4 = Cat4[(Cat4['Commodity_Description'] == 'Macadamia, Inshell Basis')
               | (Cat4['Commodity_Description'] == 'Macadamia, Wet -Inshell Basis')
               | (Cat4['Commodity_Description'] == 'Oil, Coconut')
               | (Cat4['Commodity_Description'] == 'Oil, Palm')
               | (Cat4['Commodity_Description'].str.match('Oilseed'))
               | (Cat4['Commodity_Description'] == 'Oil, Palm Kernel')].index
  
Cat4.drop(indexes4,inplace=True)

print(Cat4.Commodity_Description[Cat4.Category == 4].unique())
```

```python
# Wrangling for category 5 : Miscellanious 


Cat5 = Cat4

Cat5 = Cat5[Cat5.Commodity_Description != 'Tobacco, Unmfg., Total']

print(Cat5.Commodity_Description[Cat5.Category == 5].unique())
```

```python
# Dropping the Null Values

Cat5 = Cat5.dropna()
```

```python
# Dropping the Column not needed.

Cat5.drop(['Unnamed: 0'], axis=1)
```

```python
# Subsetting the Final dataset to only the Columns needed.

FinalUSDAProd = Cat5[["Commodity_Description", "Year","Value"]]

FinalUSDAProd
```

```python
# Renaming for further Analysis

TestingDB = pd.DataFrame(data=FinalUSDAProd)
```

```python
# Renaming the Column - 'Value' to 'Production_MT' for better readability.

TestingDB.rename(columns={'Value': 'Production_MT'}, inplace=True)
```

```python
TestingDB
```

```python
# Deleting the Commodities with missing data for years 1998 -2012

#Asparagus_Fresh                   8
#Avocados_Fresh                   11
#Fresh_Apricots                   11
#Fresh_Plums_&_Prunes             11
#Kiwifruit                         5
#Macadamia_Inshell_Basis          10
#Meat_Chicken                     14
#Pecans_Inshell_Basis             11
#Strawberries_Fresh                8


# Deleting the columns with missing data

indexesFinal = TestingDB[(TestingDB['Commodity_Description'] == 'Asparagus, Fresh')
               | (TestingDB['Commodity_Description'] == 'Avocados, Fresh')
               | (TestingDB['Commodity_Description'] == 'Fresh Apricots')
               | (TestingDB['Commodity_Description'] == 'Fresh Plums & Prunes')
               | (TestingDB['Commodity_Description'] == 'Macadamia, Inshell Basis')
               | (TestingDB['Commodity_Description'] == 'Meat, Chicken')
               | (TestingDB['Commodity_Description'] == 'Pecans, Inshell Basis')
               | (TestingDB['Commodity_Description'] == 'Strawberries, Fresh')
               | (TestingDB['Commodity_Description'] == 'Kiwifruit')].index
  
TestingDB.drop(indexesFinal,inplace=True)

print(TestingDB.Commodity_Description.unique())
```
