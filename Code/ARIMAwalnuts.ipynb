```python
import pandas as pd
import numpy as np
import pmdarima as pm
from pmdarima.arima import auto_arima
from pmdarima.arima import ADFTest
import matplotlib.pyplot as plt
from matplotlib import pyplot
from math import sqrt
from numpy import split
from numpy import array
import statsmodels
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sympy.solvers import solve
from sympy import Symbol
```

```python
# Importing dataframe
Global = pd.read_csv("C:/Users/saslo/Data_Science/DS_Portfolio/Queen_Bees/Datasets/Wrangled_Data/pivotedGlobal.csv")

Global
```

```python
# Isolating walnuts for ARIMA
justwalnuts = Global[['Year', 'Walnuts, Inshell Basis']]
justwalnuts
```

```python
# Formatting Year to date datatype 
justwalnuts['Year'] = pd.to_datetime(justwalnuts['Year'], format='%Y')
justwalnuts
```

```python
## Setting Year as index
justwalnuts.set_index('Year', inplace=True)
type(justwalnuts)
justwalnuts
```

```python
## Looking at linear plot of walnut production
justwalnuts.plot()
```

```python
### There is definitely an obvious upward trend in Walnut production, we can estimate that the data is non-stationary. Let's verify!
```

```python
## Testing our finding against Augmented Dickey Fuller test
result=adfuller(justwalnuts['Walnuts, Inshell Basis'], autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

```python
### P-value is significant confirming our data is non-stationary. We know that if we were manually fine-tuning the parameters we would have to transform the data to stationary through sequencing, but we will use the auto-arima function to do all of it for us!
```

```python
# Splitting the data for training and testing
train = justwalnuts[:20]
test = justwalnuts[-4:]
train
test
```

```python
plt.plot(train)
plt.plot(test)
```

```python
## Running auto_arima to obtain p, d, q variables
model = pm.auto_arima(justwalnuts['Walnuts, Inshell Basis'], seasonal=False)
model
```

```python
### With our seasonal parameter set to 'false', our optimal set up is p=0, d=1, q=1.
### We will set our seasonal parameter to 'true' in our training model, because that has shown in prior tests to account for variation in the prediction, rather than a simple linear prediction.
```

```python
# Manually Building Auto ARIMA model
arima_model =  auto_arima(train,start_p=0, d=1, start_q=0, 
                          max_p=1, max_d=5, max_q=5, start_P=0, 
                          D=0, start_Q=0, max_P=5, max_D=5,
                          max_Q=5, m=1, seasonal=True, 
                          error_action='warn',trace = True,
                          supress_warnings=True,stepwise = True,
                          random_state=20,n_fits = 50 )
```

```python
### We will trust in our auto_arima, and go forward with p=0, d=1, q=0
```

```python
#Summary of the model
arima_model.summary()
```

```python
# Forecasting on the test data
prediction = pd.DataFrame(arima_model.predict(n_periods = 4),index=test.index)
prediction.columns = ['predicted_values']
prediction
```

```python
# Plotting against actual values
plt.figure(figsize=(8,5))
plt.plot(train,label="Training")
plt.plot(test,label="Test")
plt.plot(prediction,label="Predicted")
plt.legend(loc = 'best')
plt.show()
```

```python
# Looking at R2 score
test['predicted_values'] = prediction
r2_score(test['Walnuts, Inshell Basis'], test['predicted_values'])
```

```python
## Not too far from 0!
```

```python
# Interpreting residual plots
arima_model.plot_diagnostics(figsize=(9,7))
plt.show()
```
