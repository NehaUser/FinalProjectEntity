```python
#Load Libraries

import pandas as pd
from scipy import stats
import numpy as np
```

```python
#Load Data

usdaHoney = pd.read_csv("../Data/FinalusdaHoney3.csv")

usdaHoney.head(50)
```

```python
TestingDB = pd.DataFrame(data=usdaHoney)


pivotedTable = TestingDB.pivot(index='Year', columns='Commodity_Description', values='Production_MT')\
            .reset_index()

pivotedTable.columns.name=None

print(pivotedTable)
```

```python
pivotedTable.count()
```

```python
# Doing correlation between all the variables and picking the ones that are highly correlated to Honey

corrMat = pivotedTable.corr()

corrMat['Honey']
```

```python
TestingDB1 = pd.DataFrame(data=corrMat['Honey'])

PosCorr_Honey = TestingDB1[TestingDB1.Honey >= 0.65]

NegCorr_Honey = TestingDB1[TestingDB1.Honey <= -0.65]

#PosCorr_Honey.append(NegCorr_Honey)

# HighCorr_Comm = pd.DataFrame(data = PosCorr_Honey)

# PosCorr_Honey.head()

display(PosCorr_Honey, NegCorr_Honey)
```
