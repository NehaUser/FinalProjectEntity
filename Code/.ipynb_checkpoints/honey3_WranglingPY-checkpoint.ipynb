```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
import numpy as np
```

```python
# Loading in dataframe
honey3 = pd.read_csv("C:/Users/saslo/Data_Science/DS_Portfolio/Queen_Bees/Datasets/Wrangled_Data/usdaHoney3.csv")
```

```python
# Dropping id column
honey3.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
honey3
```

```python
# Removing entries with NaN values
indexeshoney3 = honey3[(honey3['Commodity'] == 'Asparagus, Fresh')
                       | (honey3['Commodity'] == 'Avocados, Fresh')
                       | (honey3['Commodity'] == 'Fresh Apricots')
                       | (honey3['Commodity'] == 'Fresh Plums & Prunes')
                       | (honey3['Commodity'] == 'Macadamia, Inshell Basis')
                       | (honey3['Commodity'] == 'Meat, Chicken')
                       | (honey3['Commodity'] == 'Pecans, Inshell Basis')
                       | (honey3['Commodity'] == 'Strawberries, Fresh')
                       | (honey3['Commodity'] == 'Kiwifruit')
                       | (honey3['Commodity'] == 'Potato Products, Frozen')].index
honey3.drop(indexeshoney3, inplace=True)
                          
honey3.Commodity.unique()
```

```python
# Reshaping data for time series analysis
pivoted = honey3.pivot(index='Commodity', columns='Year', values='Metric.Tons')\
            .reset_index()
pivoted.columns.name=None
print(pivoted)
```

```python
# Looking at new dataframe
pivoted
```

```python
# Exporting pivoted df for repeated measures ANCOVA in R
pivoted.to_csv("C:/Users/saslo/Data_Science/DS_Portfolio/Queen_Bees/Datasets/Wrangled_Data/usdaHoney4_TIMEpivot.csv", index=False)
```
