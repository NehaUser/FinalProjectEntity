```r
# Separating data for conversion
Cat1CWE <- filter(Cat1, Unit_Description == '(1000 MT CWE)');
Cat1CWE
```

```r
# Converting & recoding columns
Cat1CWE <- mutate(Cat1CWE, Value = Value * 1000)
CWE <- mutate(Cat1CWE, Unit_Description = "MT")
CWE
```

```r
# Dropping previous CWE data
Cat1 <- subset(Cat1, Unit_Description != '(1000 MT CWE)');
```

```r
# Appending converted CWE Measurements to Cat1 df
Cat1 <- rbind(Cat1, CWE);
Cat1
```

```r
# Converting Cat2 value from (1000 MT) to (MT)
newmt <- mutate(Cat2, Value = Value * 1000)
newDESC <- mutate(newmt, Unit_Description = "MT")
Cat2 <- newDESC
Cat2
```
